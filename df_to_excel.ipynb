```python
#constants
IT_Power_Capacity_kw = 825
Cooling_Capacity_kw = 600
Hours_in_Year = 8760

#range
IT_Utilization_Range = [0.4, 0.9]
Cooling_Utilization_Range = [0.7, 0.9]
PUE_Range = [1.5, 1.6]

#Step size
STEP = 0.05
#Collect data
data = []

#
#iterating through ranges, step size can be adjusted
for IT_util in range(int(IT_Utilization_Range[0]*100), int(IT_Utilization_Range[1]*100)+1, int(STEP*100)):
  for Cooling_Util in range(int(Cooling_Utilization_Range[0]*100), int(Cooling_Utilization_Range[1]*100)+1, int(STEP*100)):
    PUE_values = np.arange(PUE_Range[0], PUE_Range[1]+0.1, 0.1)
    for PUE in PUE_values:
      IT_load_Mw = IT_Power_Capacity_kw*(IT_util/100)
      Cooling_load_Mw = Cooling_Capacity_kw*(Cooling_Util/100)
      total_load = IT_load_Mw*(PUE/10)
      Annual_Energy_Consumption = total_load*Hours_in_Year
      data.append((IT_util/100, Cooling_Util/100, PUE/10, IT_load_Mw, Cooling_load_Mw, total_load, Annual_Energy_Consumption))
```

```python
#creating data frame

df = pd.DataFrame(data, columns=["IT Utilization", "Cooling Utilization", "PUE", "IT Load(kw)", "Cooling Load(kw)", "Total Load(kw)", "Annual Energy consumption(KWh)"])

#write df to excel

with pd.ExcelWriter("output.xlsx", engine="openpyxl") as writer:
  df.to_excel(writer, sheet_name="sheet1", index=False)
```
